```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) # Adjusts the configuration of the plots we will create

# Read in the Data

df = pd.read_csv(r'C:\Users\Juan Songhai\Downloads\archive (3)\movies.csv')
```

```python
#Let's Look at the data
df.head()
```

```python
# Looking for missing Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
#Dropping Rows with Missing Data

df = df.dropna()
```

```python
# Looking for missing Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
# Data Types for the columns

df.dtypes
```

```python
#Changing the Columns Data Type

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df
```

```python
#Extracting Correct Year Column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Looking for any duplicates

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
#Scatter Plot Correlating Budget and Gross Rev

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budge Vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
# Plot Budget vs Gross Correlation using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "tan"}, line_kws={"color": "black"})
```

```python
#Looking Further Into Correlation
#High Correlation between Budget and Gross Revenue

df.corr()
```

```python
Correlation_matrix = df.corr(method='pearson')

sns.heatmap(Correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#Looking at Company and Gross Revenue Correlation
#Making the non numeric types of data numeric

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
#Matrix Using the New Converted Numeric Data

Correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(Correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
